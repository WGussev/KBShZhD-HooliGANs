```python
import pandas as pd
import numpy as np
```

```python
df_plan_raw = pd.read_excel('../example/Пример плана.xlsx', skiprows=4)
```

```python
df_plan_raw.columns = ['contractor', 'machine', 'empty1', 'segment', 'work', 'volume', 'empty2', 'empty3']
```

```python
df_plan_raw = df_plan_raw[['contractor', 'machine', 'segment', 'work', 'volume']].dropna()
df_plan_raw['volume_planned'] = pd.to_numeric(df_plan_raw['volume'].str.split(' ').apply(lambda x: x[0]), errors='coerce')
df_plan_raw['machine_type'] = df_plan_raw['machine'].str.split('№').apply(lambda x: x[0])
df_plan_raw['machine_number'] = df_plan_raw['machine'].str.split('№').apply(lambda x: x[1])
```

```python
df_plan_raw['volume_planned'] = pd.to_numeric(df_plan_raw['volume'].str.split(' ').apply(lambda x: x[0]), errors='coerce')
df_plan_raw['machine_type'] = df_plan_raw['machine'].str.split('№').apply(lambda x: x[0])
df_plan_raw['machine_number'] = df_plan_raw['machine'].str.split('№').apply(lambda x: x[1])
```

```python
df_plan_raw['volume_factual'] = df_plan_raw['volume_planned'] * np.random.uniform(0.9, 1.5, len(df_plan_raw))
```

```python
df_plan_raw = df_plan_raw.dropna()
```

```python
df_plan_raw = df_plan_raw.rename({'machine_type': 'machine', 'machine': 'machine_full'}, axis=1)
```

```python
df_plan_raw['machine_number'] = '№ ' + df_plan_raw['machine_number']
```
